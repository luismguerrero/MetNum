```python
import math
import numpy as np
import pylab as pl
g=9.8
m= 68.1
es= 0.0005
et=1
vel_a=0
vel_n=0
c=12.5
t=0
vin=0

velocidada= []
velocidadn= []
cont = []
for x in xrange(100):
      if et >= es:
            vel_a= (g*m/c)*(1-math.exp(-(c/m)*t))
            vel_n= vin+(g-((c/m)*vin))*((t+1)-t)
            vin=vel_n
            cont.append(t)
            t=t+1
            velocidada.append(vel_a)
            velocidadn.append(vel_n)
            if(vel_a>=1):
                et  = (vel_n-vel_a)/vel_a

            print ("Tiempo %d ") %t
            print ("Velocidad analitica: %f") %vel_a
            print ("Velocidad numerica: %f") %vel_n
            print ("Error: %f") % et          
vana= pl.plot(cont,velocidada,"r")
vnum= pl.plot(cont, velocidadn,"g")
pl.title("Velocidades")
pl.xlabel("Tiempo")
pl.ylabel("Velocidad")
pl.legend( [vana,vnum], ("Velocidad analitica", "velocidad aproximada") , "best", numpoints=1)
pl.show()
```
