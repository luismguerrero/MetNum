```python
import math
import numpy as np
import pylab as pl
valx1 = []
valy1 = []
valx = []
valy = []
relat = []
apro=[]

g=9.8
t=0
mas=68.1
c=12.5

print "t    Mtd Analitico   Mtd Numerico  Error verdadero   Error Relativo   Error Aproximado \n"

def v(t):
    return ((g*mas/c)*(1-math.exp(-(c/mas)*t)))

while (t<40):
    valx1.append(t)
    valy1.append(v(t))
    t=t+1
ti=0
aux=2
ti1=ti+aux
vt=0
ant=-1

def vmn(ti,ti1):
    return vt+(g-(c/mas)*vt)*(ti1-ti)

while (ti<40):
    valx.append(ti)
    valy.append(vt)
    verdadero=v(ti)-vt
    if(vt==0):
        relativo=(verdadero/0.000000001)*100
    
    if(vt!=0):
        relativo=(verdadero/vt)*100
    relat.append(-1*(relativo))
    
    if(vt==0):
        aproximado=((vt-ant)/0.000000001)*100
    
    if(vt!=0):
        aproximado=((vt-ant)/vt)*100
    
    if(ant<0):
        aproximado=" 0 "
        apro.append(100)
    
    if(ant>=0):
        apro.append(aproximado)
    
    print ti," ",v(ti)," ",vt, " ",verdadero," ",relativo," ",aproximado
    
    ant=vt
    vt=vmn(ti,ti1)
    ti=ti+aux
    ti1+=aux
    

f1= pl.plot(valx,relat,"b")
f2= pl.plot(valx,apro,"r")

pl.title("Paracaidista Lab3 \n ")
pl.xlabel("x")
pl.ylabel("y")
f=["Error Relativo Porcentual Verdadero","Error Relativo Porcentual Aproximado"]
pl.legend(f,loc=1)
pl.xlim(0, 40)
pl.ylim(0, 110)
pl.grid(True)

pl.show()
```
