```python
import math
from tabulate import tabulate 
import matplotlib.pyplot as plt
%matplotlib inline

xReal=-1.405
a, b, c, xl, xu, xr, x1, x2=-0.5, 2.5, 4.5, 5.0, 10.0, 0.0, 0.0, 0.0
cont, x=1, 0
table, equis, function, raices, Eaprox=[], [],[], [],[]

def cuadratica():
    x1=(-b+math.sqrt(math.pow(b,2)-4*a*c))/(2*a)
    x2=(-b-math.sqrt(math.pow(b,2)-4*a*c))/(2*a)
    if(x1>x2):
        xMayor=x1
        return xMayor
    else:
        xMayor=x2
        return xMayor

def funcion(x):
    fx=a*math.pow(x,2)+b*x+c
    return fx

while (x<10):
    equis.append(x)
    fxG=a*math.pow(x,2)+b*x+c
    function.append(fxG)
    x+=0.8
    
for i in range(0,4):
    xr=(xl+xu)/2
    raices.append(xr)
    fA=funcion(xl)
    fB=funcion(xr)
    erpv=math.fabs(((cuadratica()-xr)/cuadratica())*100)
    if(i==0):
        Eaprox.append("-----")
        eaprx=100
    else:
        j=i-1
        eaprx=math.fabs(((raices[i]-raices[j])/raices[i])*100)
        Eaprox.append(eaprx)
    table.append([i+1,xr, str(erpv)+" %",str(Eaprox[i])+" %"])
    if ((fA*fB)<0):
        xu=xr
    if((fA*fB)>0):
         xl=xr

print "La raiz mayor es: x= %f\n"%(cuadratica())

print tabulate(table, headers=["Iteracion", "Raiz", "Error rela porc verd", "Error aprox"], tablefmt="pipe")

plt.title("Raiz de la funcion graficamente")
plt.axis([0, 10, -2, 10])
plt.ylabel('f(x)')
plt.xlabel('x')
plt.axhline(0, color="black")
plt.plot(equis, function, 'm-', marker='*')
plt.plot(cuadratica(),0, 'cs', label='Raiz =6.405125')
plt.legend(loc = 'upper right', numpoints = 1)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt 
from tabulate import tabulate
%matplotlib inline

x=-4
xl=1.0
xu=3.0
xr=0.0
valores, funcion, tabla, raices, tabla, Eaprox, tabla1=[],[],[],[],[],[],[]
i=0
true=2.69556

while(x<=4):
    if(x==0):
        continue
    else:
        fx=(0.8-0.3*x)/x
        valores.append(x)
        funcion.append(fx)
        x+=0.1
    tabla.append([valores[i], funcion[i]])
    i+=1

def metodo1(dato):
    fn=(0.8-0.3*dato)/dato
    return fn

b=0
while (b<3):
    f1=metodo1(xl)
    f2=metodo1(xu)
    xr=xu-(f2*(xl-xu))/(f1-f2)
    raices.append(xr)
    f3=metodo1(xr)
    erpv=math.fabs(((true-xr)/true)*100)
    if(b==0):
        Eaprox.append("-----")
        eaprx=100
    else:
        k=b-1
        eaprx=math.fabs(((raices[b]-raices[k])/raices[b])*100)
        Eaprox.append(eaprx)
    tabla1.append([b+1,xr, str(erpv)+" %", str(Eaprox[b])+" %"])
    if ((f1*f3)<0):
        xu=xr
    if((f1*f3)>0):
         xl=xr  
    b+=1
    
plt.title("f(x)")
plt.axis([-4, 5, -6, 6])
plt.ylabel('f(x)')
plt.xlabel('x')
plt.axhline(0, color="black")
plt.plot(valores, funcion, 'r-', linewidth=2)
plt.plot(true,0, 'co', label='Raiz =2.69556')
plt.legend(loc = 'upper right', numpoints = 1)
plt.show()

print "\nManera analitica :\n"
print tabulate(tabla, headers=["x     ","f(x)     "], tablefmt="pipe")
print "\n Falsa posición con 3 iteraciones :\n"
print tabulate(tabla1, headers=["Iteracion","Raiz","Error porc verd","Error aprox"], tablefmt="pipe")
```

```python
import math
from tabulate import tabulate
import matplotlib.pyplot as plt
%matplotlib inline

g, c, v, t=9.8, 15.0, 35.0, 9.0
Es=0.1
f1,f2,f3=1.0,1.0,1.0
table, valores, funcion, tabla, Eaprox, raices=[],[],[],[],[],[]
xVerd=59.8523
xl=55.0
xu=60.0
xr=0.0
Es=0.001

for i in range(-10, 200):
    if(i==0):
        continue
    m=i
    fmasa=((g*m)/c)*(1-math.pow(math.e,-(c/m)*t))-v
    if(m>0 and m%4==0):
        valores.append(m)
        funcion.append(fmasa)
        #tabla.append([m, fmasa])
        
def resultado(m):
    fm=((g*m)/c)*(1-math.pow(math.e,(-c/m)*t))-v
    return fm

b=0
while 1:
    f1=resultado(xl)
    f2=resultado(xu)
    xr=xu-(f2*(xl-xu))/(f1-f2)
    raices.append(xr)
    f3=resultado(xr)
    erpv=math.fabs(((xVerd-xr)/xVerd)*100)
    if(b==0):
        Eaprox.append("-----")
        eaprx=100
    else:
        k=b-1
        eaprx=math.fabs(((raices[b]-raices[k])/raices[b])*100)
        Eaprox.append(eaprx)
    table.append([xr, str(erpv)+" %", str(Eaprox[b])+" %"])
    if ((f1*f3)<0):
        xu=xr
    if((f1*f3)>0):
         xl=xr 
    if(eaprx<Es):
        break
    b+=1   
plt.title("MASA")
plt.axis([0, 250, -50, 50])
plt.ylabel('f(m)')
plt.xlabel('m')
plt.axhline(0, color="black")
plt.plot(valores, funcion, 'r-', linewidth=2)
plt.plot(xVerd,0, 'gs', label='Raiz =59.8523')
plt.legend(loc = 'upper right', numpoints = 1)
plt.show()

print "\n Metodo de Falsa Posicion:\n"
print tabulate(table, headers=["f(m)","Error porc verd","Error aprox"], tablefmt="pipe")
```
